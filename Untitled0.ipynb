```python
import numpy as np
import pandas as pd
import statistics as st
import scipy.stats as sp

# Getting the dataset
data = [15, 28, 99, 86, 88, 110, 103, 87, 94, 78, 77, 85, 46, 35, 68]

# testing which library has all 3 functions mean, median and mode
try:
    print("Using Numpy for Mean, Median Mode: ")
    print("Mean : ", np.mean(data))
    print("Median : ", np.median(data))
    # np doesn't have mode function directly
    print("Mode : ", sp.mode(data)[0][0])
    print("Numpy has all three!")
except Exception as e:
    print("Numpy does not have all three: ", e)

try:
    print("Using Pandas for Mean, Median Mode: ")
    print("Mean : ", pd.Series(data).mean())
    print("Median : ", pd.Series(data).median())
    print("Mode : ", pd.Series(data).mode()[0])
    print("Pandas has all three!")
except Exception as e:
    print("Pandas does not have all three: ", e)

try:
    print("Using Statistics for Mean, Median Mode: ")
    print("Mean : ", st.mean(data))
    print("Median : ", st.median(data))
    print("Mode : ", st.mode(data))
    print("Statistics has all three!")
except Exception as e:
    print("Statistics does not have all three: ", e)

# minimum, maximum, standard deviation, percentile, sort
print("Maximum: ", max(data))
print("Minimum: ", min(data))
print("Standard Deviation: ", st.stdev(data))
print("25th Percentile: ", np.percentile(data, 25))
data.sort()
print("Sorted: ", data)
print("Skewness: ", sp.skew(data))
print("Kurtosis: ", sp.kurtosis(data))
print("Range of Data: ", np.ptp(data))
print("Variance: ", np.var(data))
```
